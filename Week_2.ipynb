***Day 1***

```python
# Match Case
x = int(input("Enter the value of number: "))
match x:
  case 0:
    print("Number is 0")
  case 4:
    print("Number is 4")
  case _ if x != 50:
    print(x, "is not 50")
  case _:
    print(x)
```

```python
# For Loop
fruits = {"Apple", "Orange", "Papaya", "Banana"}
for i in fruits:
  print(fruits)
  for o in i:
    print(o)
```

```python
for k in range(1, 12, 3):
  print(k)
```

```python
# Sum upto n terms
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total = total + i
print("Sum =", total)
```

```python
# Calculating Factorials
n = int(input("Enter a number: "))
fact = 1
for i in range(1, n + 1):
    fact = fact * i
print("Factorial =", fact)
```

```python
# While Loop
c = int(input("Enter n: "))
i = 0
while i <= c  :
  print(i)
  i = i + 1
```

```python
# Reverse with While Loop
n = int(input("Enter n: "))
count = n
while count > -1:
  print(count)
  count = count - 1
```

```python
# Break
for i in range(12):
  if i == 10:
    break
  print( "5 X", i + 1, "=", 5 * (i+1))
```

```python
# Print with condition using Break & Continue
for i in range(1, 50):
  if i % 5 == 0:
    continue
  if i > 40:
    break
  print(i)
```

```python
# Using def to define function
def calculateGmean(a, b):
  mean = (a * b) / (a + b)
  print(f"{mean:.3f}")
a = float(input("Enter a: "))
b = float(input("Enter b: "))
calculateGmean(a, b)

c = float(input("Enter c: "))
d = float(input("Enter d: "))
calculateGmean(c, d)
```

```python
def isGreater(a, b):
  if a > b:
    print("The first no. is Greater")
  else:
    print("The second no. is Greater")
a = float(input("Enter a: "))
b = float(input("Enter b: "))
isGreater(a, b)

c = float(input("Enter c: "))
d = float(input("Enter d: "))
isGreater(c, d)
```

```python
# Check is a num is prime or not
def isPrime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
k = int(input("Enter: "))
if isPrime(k):
    print("Prime")
else:
    print("Not Prime")
```

```python
# Sum of digits of a num
def isSum(n):
    total = 0
    while n > 0:
        m = n % 10
        total = total + m
        n = n // 10
    print(total)
k = int(input("Enter: "))
isSum(k)
```
