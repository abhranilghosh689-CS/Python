***Day 1***

```python
# Match Case
x = int(input("Enter the value of number: "))
match x:
  case 0:
    print("Number is 0")
  case 4:
    print("Number is 4")
  case _ if x != 50:
    print(x, "is not 50")
  case _:
    print(x)
```

```python
# For Loop
fruits = {"Apple", "Orange", "Papaya", "Banana"}
for i in fruits:
  print(fruits)
  for o in i:
    print(o)
```

```python
for k in range(1, 12, 3):
  print(k)
```

```python
# Sum upto n terms
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total = total + i
print("Sum =", total)
```

```python
# Calculating Factorials
n = int(input("Enter a number: "))
fact = 1
for i in range(1, n + 1):
    fact = fact * i
print("Factorial =", fact)
```

***Day 2***

```python
# Reverse Loop
n = int(input("Enter n: "))
for k in range(n, 0, -1):
  print(k)
```

```python
# Print no. of stars
n = int(input("Enter n: "))

for k in range(n, 0, -1):
    print("*" * k)
```

```python
# Print how many no. are divisible by 3 till n
n = int(input("Enter n: "))
count = 0
for k in range(3, n, 3):
    count = count + 1

print("Count =", count)
```

***Day 3***

```python
# While Loop
c = int(input("Enter n: "))
i = 0
while i <= c  :
  print(i)
  i = i + 1
```

```python
# Reverse with While Loop
n = int(input("Enter n: "))
count = n
while count > -1:
  print(count)
  count = count - 1
```

```python
# Break
for i in range(12):
  if i == 10:
    break
  print( "5 X", i + 1, "=", 5 * (i+1))
```

```python
# Print with condition using Break & Continue
for i in range(1, 50):
  if i % 5 == 0:
    continue
  if i > 40:
    break
  print(i)
```
