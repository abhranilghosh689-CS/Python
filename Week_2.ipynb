```python
# Match Case
x = int(input("Enter the value of number"))
match x:
  case 0:
    print("Number is 0")
  case 4:
    print("Number is 4")
  case _ if x != 50:
    print(x, "is not 50")
  case _:
    print(x)
```

```python
# For Loop
fruits = {"Apple", "Orange", "Papaya", "Banana"}
for i in fruits:
  print(fruits)
  for o in i:
    print(o)
```

```python
for k in range(1, 12, 3):
  print(k)
```

```python
# Sum upto n terms
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
    total = total + i
print("Sum =", total)
```
